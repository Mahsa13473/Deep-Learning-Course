```python
# helper functions

from tensorflow import set_random_seed
set_random_seed(0)
np.random.seed(0)

# helper function for geterating the data
def data_generator(N = 200,D = 2,K = 2):
    """
    N: number of points per class; 
    D: dimensionality; 
    K: number of classes
    """
    np.random.seed(0)
    X = np.zeros((N*K,D))
    y = np.zeros((N*K), dtype='uint8')
    
    for j in range(K):
        ix = range(N*j,N*(j+1))
        r = np.linspace(0.0,1,N) # radius
        t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.3 # theta
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        y[ix] = j
    
    fig = plt.figure()
    plt.title('Figure 1: DataSet')
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
   
    plt.xlim(X.min()-.5, X.max()+.5)
    plt.ylim(X.min()-.5, X.max()+.5)
    
    return X,y


# helper function for visualizing the decision boundaries
def visualize(sample, target, model):
    """
    function for visualizing the classifier boundaries on the TOY dataset.

    sample: Training data features
    target: Target 
    model: the model
    """
    h = 0.02
    x_min, x_max = sample[:, 0].min() - 1, sample[:, 0].max() + 1
    y_min, y_max = sample[:, 1].min() - 1, sample[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    

    Z = model.predict_classes(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    fig = plt.figure()
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
    plt.scatter(sample[:, 0], sample[:, 1], c=target, s=40, cmap=plt.cm.Spectral)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    
    return
    
```

```python
#  TOY DataSet
sample, target = data_generator(N = 200)
# print(target.shape)
```

Compute and visualize the equation of the decision Boundary 

```python
# weights = logistic_regression.get_weights()
# print(weights)

# helper function for visualizing the decision boundaries
# implement a new function to draw the decision boundary by its equation, not the Monte Carlo method. 
def visualize_line(sample, target, model):
    """
    function for visualizing the classifier boundaries on the TOY dataset.

    sample: Training data features
    target: Target 
    model: the model
    """
    h = 0.02
    x_min, x_max = sample[:, 0].min() - 1, sample[:, 0].max() + 1
    y_min, y_max = sample[:, 1].min() - 1, sample[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))


    Z = model.predict_classes(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    threshold = 0.5
    Z_score = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z_score = Z_score.reshape(xx.shape)
    Z_score = Z_score - threshold

    eps = 0.02
    
    # Find two different points on the decision boundary 
    for i in range(xx.shape[0]):
        for j in range(xx.shape[1]):
            
            if abs(Z_score[i][j])< eps:
                p1_x, p1_y = xx[i][j], yy[i][j]
                break
                    
                    
    for i in range(xx.shape[0]-1, 0, -1):
        for j in range(xx.shape[1]-1, 0, -1):
            
            if abs(Z_score[i][j])< eps:
                p2_x, p2_y = xx[i][j], yy[i][j]
                break

    
    # print("points:", "(", p1_x, p1_y,"), (", p2_x, p2_y, ")")
    line_equation(p1_x, p1_y, p2_x, p2_y)

            
    fig = plt.figure()
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
    plt.scatter(sample[:, 0], sample[:, 1], c=target, s=40, cmap=plt.cm.Spectral)
    
    x1, y1 = [p1_x, p2_x], [p1_y, p2_y]
    
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.plot(x1, y1, color='green')
    
    return


# line equation based on two points on the line
def line_equation(p1_x, p1_y, p2_x, p2_y):
    m = (p2_y-p1_y)/(p2_x-p1_x)
    b = p1_y - m*p1_x
    
    print("decision boundary equation:", "x2 =", round(m, 2), "* x1 +", round(b, 2))
    

visualize_line(sample,target, logistic_regression)
```

5. Add L2-regularization to your implemented nonlinear neural network in (4.). Set the coefficient of L2-regularization to be 0.01, 2, 100, respectively. How do different values of coefficient of L2-regularization affect the model (i.e., model parameters, loss value, accuracy, decision boundary)? You can use a table to compare models trained without regularization, with different coefficients of regularization. (20 points) 
    * Please draw your table and analysis in the '**Answers and Analysis**' section.

```python
def net(reg_lambda):
    model = keras.Sequential()
    model.add(keras.layers.Dense(10, activation='relu', kernel_regularizer=keras.regularizers.l2(l=reg_lambda)))
    model.add(keras.layers.Dense(10, activation='relu', kernel_regularizer=keras.regularizers.l2(l=reg_lambda)))
    model.add(keras.layers.Dense(1, activation='sigmoid'))
    
    return model
```

$\lambda=0.01:$

```python
# Early Stopping
callback = tf.keras.callbacks.EarlyStopping(monitor='loss',min_delta=0, patience=3)

#Define the model
DNN_reg_01 = net(0.01)

optimizer = keras.optimizers.Adam(lr=0.001)
# optimizer = keras.optimizers.SGD(lr=0.001)

DNN_reg_01.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])


# Train the model
value_epoches = DNN_reg_01.fit(sample, target, epochs=100, batch_size=1) #, callbacks=[callback]) 

eval = DNN_reg_01.evaluate(sample, target)
```
